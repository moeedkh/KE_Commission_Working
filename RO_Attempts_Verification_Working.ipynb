```python
# Add relevant public holidays here
import pandas as pd
from pandas.tseries.offsets import BDay
from pandas.tseries.offsets import CustomBusinessDay
from pandas.tseries.holiday import AbstractHolidayCalendar, Holiday
import numpy as np

# Define the public holidays for Pakistan (example for a given year)
class PakistanHolidays(AbstractHolidayCalendar):
    rules = [
        # Kashmir Day
        Holiday('Kashmir Day', month=2, day=5),
        # Eid-ul-Fitr example (in reality, Eid dates change each year based on the Islamic calendar)
        Holiday('Eid-ul-Fitr_Day1', month=3, day=31),
        # Eid-ul-Adha example
        Holiday('Eid-ul-Fitr_Day2', month=4, day=1),
        # Pakistan's Independence Day
        Holiday('Independence Day', month=8, day=14),
        # Eid Milad-un-Nabi (example)
        Holiday('Eid Milad-un-Nabi', month=9, day=5),
        # Other holidays can be added here
    ]



```

```python
from geopy.distance import geodesic

# Function to calculate distance
def calculate_distance(row):
    # Check for NaN values in the coordinates
    if pd.isna(row['Longitude']) or pd.isna(row['Latitude']) or pd.isna(row['KE_long']) or pd.isna(row['KE_lat']):
        return None  # Or return a default value, like 0 or np.nan if you prefer
    location1 = (row['Longitude'], row['Latitude'])
    location2 = (row['KE_long'], row['KE_lat'])
    return geodesic(location1, location2).meters

# Apply the function to your dataframe
App['distance_meters'] = App.apply(calculate_distance, axis=1)

# Show the DataFrame with the calculated distances
App.head()

```

```python
from fuzzywuzzy import fuzz
import pandas as pd

# Define full form and short form remarks
full_remarks = {
    "AGREED TO PAY": ["AGREED TO PAY", "ATP"],
    "BILL NOT RECEIVED": ["BILL NOT RECEIVED", "BNR"],
    "BILLING DISPUTE": ["BILLING DISPUTE", "BD"],
    "COURT CASE": ["COURT CASE", "CC"],
    "FOLLOW UP VISIT REQUIRED": ["FOLLOW UP VISIT REQUIRED", "FUV"],
    "NEGOTIATIONS UNDER WAY": ["NEGOTIATIONS UNDER WAY", "NUW"],
    "WARNING": ["WARNING", "W"],
    "ALREADY PAID": ["ALREADY PAID", "AP"]
}

# Function to categorize remarks based on fuzzy matching
def categorize_remarks(remark):
    # Convert to uppercase to make comparison case-insensitive
    remark = remark.upper()  # Optional but helps with case sensitivity

    for full_form, short_forms in full_remarks.items():
        for form in short_forms:
            # Use fuzzy matching to compare the remark with the short form and full form
            if fuzz.partial_ratio(remark, form) > 80:  # You can adjust the threshold (e.g., 80)
                return full_form
    return "Not_Eligible"  # Return "UNKNOWN" if no match is found

# Apply the categorization function to the 'KERemarks' column
LST_merged['Final_Remarks'] = LST_merged['KERemarks'].apply(categorize_remarks)

# Display the updated DataFrame with the categorized remarks
print(LST_merged[['KERemarks', 'Final_Remarks']].head())
```
